```python
###What are the top 5 most installed apps in  each popular category
```

```python
df.head()
```

```python
dfa=df.groupby(["Category","App"])["Installs"].sum().reset_index()
```

```python
dfa
```

```python
dfa.sort_values("Installs",ascending=False)
```

```python
dfa=dfa.sort_values("Installs",ascending=False)
```

```python
dfa[dfa.Category=='GAME'][:5]
```

```python
dfa[dfa.Category=='PRODUCTIVITY'][:5]
```

```python
apps=["GAME","PRODUCTIVITY","SOCIAL","COMMUNICATION"]
sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):
    df2=dfa[dfa.Category==app]
    df3=df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data=df3,x="Installs",y="App")
    plt.x_label("Installation in millions")
    plt.y_label("")
    plt.title(app,size=20)
plt.tight_layout()
plt.subpplots_adjust(hspace=.3)
plt.show()
```
